```python
# prompt: Write functions to create visualizations such as histograms, box plots, and scatter plots to explore the distribution and relationships between different variables in the sales data. Use matplotlib or seaborn libraries to create these visualizations for above program

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "/content/sample_data/sales_data_sample (1).csv"

def summary_statistics(file_path):
    """
    Calculates and displays summary statistics for key columns in sales data.

    Args:
        file_path (str): The path to the CSV file containing the sales data.
    """
    try:
        sales_data = pd.read_csv(file_path)  # Assuming the file is a CSV
        # Select relevant numerical columns for summary statistics
        numerical_cols = sales_data.select_dtypes(include=['number']).columns

        if not numerical_cols.empty:  #Check if there are any numerical columns
            summary = sales_data[numerical_cols].describe()
            print(summary)
        else:
            print("No numerical columns found in the dataset.")

    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
    except pd.errors.ParserError:
        print(f"Error: Could not parse the CSV file '{file_path}'. Check the file format.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

def visualize_sales_data(file_path):
    """
    Creates visualizations for sales data, including histograms, box plots, and scatter plots.

    Args:
        file_path (str): The path to the CSV file containing sales data.
    """
    try:
        sales_data = pd.read_csv(file_path)
        numerical_cols = sales_data.select_dtypes(include=['number']).columns

        if not numerical_cols.empty:
            # Histograms
            for col in numerical_cols:
                plt.figure(figsize=(8, 6))
                sns.histplot(sales_data[col], kde=True)
                plt.title(f'Distribution of {col}')
                plt.xlabel(col)
                plt.ylabel('Frequency')
                plt.show()

            # Box plots
            plt.figure(figsize=(10, 6))
            sns.boxplot(data=sales_data[numerical_cols])
            plt.title('Box Plots of Numerical Columns')
            plt.xticks(rotation=45)
            plt.show()

            # Scatter plots (example with two numerical columns - adjust as needed)
            if len(numerical_cols) >= 2:
                plt.figure(figsize=(8, 6))
                sns.scatterplot(x=numerical_cols[0], y=numerical_cols[1], data=sales_data)
                plt.title(f'Scatter Plot of {numerical_cols[0]} vs. {numerical_cols[1]}')
                plt.xlabel(numerical_cols[0])
                plt.ylabel(numerical_cols[1])
                plt.show()
        else:
            print("No numerical columns found in the dataset for visualization.")

    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
    except pd.errors.ParserError:
        print(f"Error: Could not parse the file '{file_path}'. Check the file format.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")


# Example usage
# summary_statistics(file_path)
visualize_sales_data(file_path)
```

```python
# prompt: Write a main function called main that orchestrates the execution of the above functions. The main function should load the sales data, call the summary_statistics function to display summary statistics, and call the data visualization functions to create visualizations.

def main():
    """Orchestrates the execution of data analysis functions."""
    file_path = "/content/sample_data/sales_data_sample (1).csv"  # Replace with the actual file path

    try:
        print("Displaying Summary Statistics:")
        summary_statistics(file_path)
        print("\nCreating Data Visualizations:")
        visualize_sales_data(file_path)

    except Exception as e:
        print(f"An error occurred in the main function: {e}")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Function to calculate summary statistics
def summary_statistics(data):
    summary = data.describe()
    print("Summary Statistics:")
    print(summary)

# Function to create histogram
def create_histogram(data, column):
    plt.figure(figsize=(8, 6))
    if column in data.columns:
        sns.histplot(data[column], bins=20, kde=True)
        plt.title(f'Histogram of {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.show()
    else:
        print(f"Error: Column '{column}' not found. Available columns: {data.columns.tolist()}")

# Function to create box plot
def create_boxplot(data, column):
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=data[column])
    plt.title(f'Box Plot of {column}')
    plt.xlabel(column)
    plt.show()

# Function to create scatter plot
def create_scatterplot(data, x_column, y_column):
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=data[x_column], y=data[y_column])
    plt.title(f'Scatter Plot of {x_column} vs {y_column}')
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.show()

# Main function
def main():
    # Load data from CSV file
    sales_data = pd.read_csv("/content/sample_data/sales_data_sample (1).csv", encoding='latin-1')

    # Display summary statistics
    summary_statistics(sales_data)

    # Create histogram of quantity sold
    create_histogram(sales_data, 'quantity_sold')

    # Create box plot of price per unit
    create_boxplot(sales_data, 'price_per_unit')

    # Create scatter plot of quantity sold vs price per unit
    create_scatterplot(sales_data, 'quantity_sold', 'price_per_unit')

if __name__ == "__main__":
    main()
```
