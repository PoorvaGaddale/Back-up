```python
# prompt: Using sample sales data for above which includes information such as product ID, quantity sold, price per unit, transaction ID, customer name, and purchase date.

import pandas as pd

# Function to identify outliers
def identify_outliers(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    return outliers

    # Function to display outliers
def display_outliers(outliers):
    print("Outliers:")
    print(outliers)

cgpa_column = 'cgpa'
placement_exam_marks_column = 'placement_exam_marks'

def main():
# Load data from CSV file
  sales_data = pd.read_csv('/content/sample_data/placement.csv')
  print(sales_data.head())
  column_names = sales_data.columns
  print(column_names)

  quantity_outliers = identify_outliers(sales_data[cgpa_column])
  display_outliers(quantity_outliers)

  price_outliers = identify_outliers(sales_data[placement_exam_marks_column])
  display_outliers(price_outliers)



if __name__ == "__main__":
    main()
```

```python
# prompt: Draw box plot graph for above outliers

import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'sales_data' DataFrame is already loaded and outliers are identified
# as 'quantity_outliers' and 'price_outliers'

def main():
    sales_data = pd.read_csv('/content/sample_data/placement.csv')
    cgpa_column = 'cgpa'
    placement_exam_marks_column = 'placement_exam_marks'

    quantity_outliers = identify_outliers(sales_data[cgpa_column])
    price_outliers = identify_outliers(sales_data[placement_exam_marks_column])

    plt.figure(figsize=(10, 6))  # Adjust figure size as needed

    # Box plot for cgpa outliers
    plt.subplot(2, 1, 1)
    plt.boxplot(sales_data[cgpa_column], vert=False)  # vert=False for horizontal boxplot
    plt.title('Box Plot of CGPA with Outliers')
    plt.xlabel('CGPA')

    # Mark the outliers
    plt.scatter(quantity_outliers, [1]*len(quantity_outliers), color='red', marker='o', label='Outliers') # Assuming single row for horizontal boxplot
    plt.legend() # Show the legend for outliers

    # Box plot for placement exam marks outliers
    plt.subplot(2, 1, 2)
    plt.boxplot(sales_data[placement_exam_marks_column], vert=False)  # vert=False for horizontal boxplot
    plt.title('Box Plot of Placement Exam Marks with Outliers')
    plt.xlabel('Placement Exam Marks')

    # Mark the outliers
    plt.scatter(price_outliers, [1]*len(price_outliers), color='red', marker='o', label='Outliers') # Assuming single row for horizontal boxplot
    plt.legend() # Show the legend for outliers

    plt.tight_layout()  # Adjust layout to prevent overlapping
    plt.show()

# ... (rest of your code for identify_outliers and display_outliers)

if __name__ == "__main__":
    main()
```
